```python
# Word clouds are commonly used to perform high-level analysis and visualization of text data.
# Project: Word Cloud Analysis of Alices Adventures in Wonderland!

#Step 1: import Libraries.
# Step 1.1: Install the WordCloud library
!pip install wordcloud

# Step 1.2: Import WordCloud class and list of stopwords
from wordcloud import WordCloud,STOPWORDS
print('WordCloud imported')

#Step 1.3: Import the urllib library
import urllib

#Step 1.4: Import Matplotlib for displaying the word cloud
import matplotlib.pyplot as plt


#Step 2: Download the Alice in Wonderland text file from the web
alice_novel = urllib.request.urlopen('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/alice_novel.txt').read().decode("utf-8")

#Step 3: Create a set of common stopwords to remove from the text
stopwords=set(STOPWORDS)

#Step 4: # (Optional) If you get an AttributeError about image/font, upgrade these:
%pip install --upgrade Pillow
%pip install --upgrade numpy

#Step 5: Instantiate a WordCloud object
#For simplicity, let's generate a word cloud using only the first 2000 words in the novel.
alice_wc=WordCloud()
    
#Step 6: Create the word cloud using the novel text
alice_wc.generate(alice_novel)

#Step 7: Display the generated word cloud
    
# Step 7.1: Display the word cloud image smoothly using bilinear interpolation
plt.imshow(alice_wc, interpolation ='bilinear')
    
# Step 7.2: Hide the x and y axes
plt.axis('off')

# Step 7.3: Display the final word cloud output.
plt.show()


```

```python
#Step 8: Create a new figure and set its size and dispy.

#8.1: Create a new figure and set its size
fig=plt.figure(figsize=(14,18))

#8.2: Show word cloud
plt.imshow(alice_wc, interpolation='bilinear')

#8.3: Hide the axis lines
plt.axis('off')

#8.4: Display the image
plt.show()

```

```python
#Step 9: Add 'said' to stopwords to remove it from the cloud (since its not informative)
    
#9.1: add the words said to stopwords
stopwords.add('said')

#9.2: Re-create the WordCloud with the updated stopwords
alice_wc=WordCloud(stopwords=stopwords)

#9.3: Generate the word cloud again
alice_wc.generate(alice_novel)

#9.4:Create a new figure and set its size.
fig = plt.figure(figsize=(14, 18))

#9.5: Show word cloud
plt.imshow(alice_wc, interpolation='bilinear')

#9.6: Hide the axis lines
plt.axis('off')

#9.7: Display the image
plt.show()

```

```python
#Next step is very creative and fun! Were going to shape our word cloud into a picture  in this case, Alice and her rabbit 

# Step 10: Load and convert the Alice mask image for shaping the word cloud.
# 10.1: Import required libraries
import numpy as np                     # for numerical operations
from PIL import Image                  # for image handling
import urllib                          # for downloading from a URL

#10.1: Load and convert the Alice mask image for shaping the word cloud

alice_mask = np.array(Image.open(urllib.request.urlopen('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/Module%204/images/alice_mask.png')))
print("Alice mask image loaded successfully!")
```

```python
#Step 11: display this Alice mask image

#11.1: Create a new figure and set its size.
fig=plt.figure(figsize=(14,18))

#11.2: Show the Alice mask image
plt.imshow(alice_mask,cmap=plt.cm.gray, interpolation='bilinear')

#11.3: Hide the axis lines
plt.axis('off')

#11.4: Display the image
plt.show()
```

```python
#Step 12: Generate and Display the Alice-shaped Word Cloud Using the Mask.

# 12.1: instantiate a word cloud object
alice_wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask, stopwords=stopwords)

#12.2:  generate the word cloud
alice_wc.generate(alice_novel)

#12.3: display the word cloud
fig = plt.figure(figsize=(14, 18))

#12.4: Show word cloud.
plt.imshow(alice_wc, interpolation='bilinear')

#12.5: Hide the axis lines.
plt.axis('off')

#12.6: Display the image
plt.show()
```
